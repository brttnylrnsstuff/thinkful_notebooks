```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Returns plots inline without saving to files
%matplotlib inline
```

```python
df = pd.DataFrame()

df['x'] = np.random.normal(0, 20, 1000)
df['y'] = np.random.normal(20, 5, 1000)
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221, facecolor='gray')
plt.hist(df['x'], bins=40, color='green')
plt.hist(df['y'], bins=40, color='blue')
plt.title('A Couple Histograms')
plt.xlabel('Random Numbers')
plt.ylabel('Occurances')

plt.subplot(222)
plt.scatter(df['x'],df['y'])

plt.subplot(223)
plt.boxplot(df['x'])
plt.title('A Box Plot')
plt.ylabel('Random Numbers')

plt.subplot(224)
plt.boxplot(df['y'])
plt.title('Another Box Plot')
plt.ylabel('Random Numbers')

plt.show()
```
