```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.DataFrame()
data['Gender'] = ['Male'] * 1000 + ['Female'] * 1000
data['Height'] = np.append(np.random.normal(69, 8, 1000), np.random.normal(64, 5, 1000))
data['Weight'] = np.append(np.random.normal(195, 25, 1000), np.random.normal(166, 15, 1000))
```

```python
plt.figure(figsize=(15,5))

ax = plt.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Weight', fontsize=16)
plt.ylabel('Count', fontsize=16)

plt.hist(data['Weight'], color='#3F5D7D', bins=100)
plt.axvline(data['Weight'].mean(), color='red', linestyle='solid', linewidth=3)
plt.axvline(data['Weight'].median(), color='green')
plt.axvline(np.percentile(data['Weight'], 5), color='red')
plt.axvline(np.percentile(data['Weight'], 10), color='orange')
plt.axvline(np.percentile(data['Weight'], 90), color='blue')
plt.axvline(np.percentile(data['Weight'], 95), color='purple')

#plt.savefig("rando-histo.png", bbox_inches="tight")
plt.show()
```

```python
values, base = np.histogram(data['Weight'], bins=2000)
cumulative = np.cumsum(values)

plt.figure(figsize=(15,5))

ax1 = plt.subplot(111)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.get_xaxis().tick_bottom()
ax1.get_yaxis().tick_left()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Weight', fontsize=16)
plt.ylabel('Count', fontsize=16)

plt.plot(base[:-1], cumulative, c='blue')
plt.axhline(0.05 * len(cumulative), color='red')
plt.axhline(0.10 * len(cumulative), color='orange')
plt.axhline(0.50 * len(cumulative), color='green')
plt.axhline(0.90 * len(cumulative), color='blue')
plt.axhline(0.95 * len(cumulative), color='purple')
plt.axvline(data['Weight'].mean(), color='red', linewidth='3')
plt.axvline(data['Weight'].median(), color='green')
plt.axvline(np.percentile(data['Weight'], 5), color='red')
plt.axvline(np.percentile(data['Weight'], 10), color='orange')
plt.axvline(np.percentile(data['Weight'], 90), color='blue')
plt.axvline(np.percentile(data['Weight'], 95), color='purple')

plt.show()
```
