```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Returns plots inline without saving to files
%matplotlib inline
```

```python
np.random.seed(1221)
df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.scatter(
    x=df['rand'],
    y=df['rand_sq'],
    color='purple',
    marker='x', s=10
)
plt.xlabel('rand')
plt.ylabel('rand_sq')
plt.title('X Squared')

plt.tight_layout()
plt.show()
```
